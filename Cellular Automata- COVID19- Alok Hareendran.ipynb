```python
import random
import numpy as np
import matplotlib.pyplot as plt

lst=[]
prob = [0.95,0.8,0.67,0.58,0.5,0.3]
reco=[0.01,0.1,0.16,0.2,0.24,0.28]
for i in range(0,len(prob)):
    INFECTION_PROB = prob[i]
    RECOVERY_PROB = reco[i]
    # Constants
    GRID_WIDTH = 100
    GRID_HEIGHT = 100
    INITIAL_INFECTED_PROB = 0.01
    NUM_TIME_STEPS = 25

    # States of a cell in the grid
    HEALTHY = 0
    INFECTED = 1
    RECOVERED = 2

    def initialize_grid():
        grid = np.zeros((GRID_HEIGHT, GRID_WIDTH))
        for y in range(GRID_HEIGHT):
            for x in range(GRID_WIDTH):
                if random.random() < INITIAL_INFECTED_PROB:
                    grid[y][x] = INFECTED
        return grid

    def count_infected_neighbors(grid, x, y):
        count = 0
        for i in range(-1, 2):
            for j in range(-1, 2):
                if not (i == 0 and j == 0):
                    if 0 <= y + i < GRID_HEIGHT and 0 <= x + j < GRID_WIDTH:
                        if grid[y + i][x + j] == INFECTED:
                            count += 1
        return count

    def update_grid(grid):
        new_grid = np.copy(grid)
        for y in range(GRID_HEIGHT):
            for x in range(GRID_WIDTH):
                if grid[y][x] == HEALTHY:
                    if count_infected_neighbors(grid, x, y) > 0 and random.random() < INFECTION_PROB:
                        new_grid[y][x] = INFECTED
                elif grid[y][x] == INFECTED:
                    if random.random() < RECOVERY_PROB:
                        new_grid[y][x] = RECOVERED
                elif grid[y][x] == RECOVERED:
                    if count_infected_neighbors(grid, x, y) > 2 and random.random() < INFECTION_PROB*0.4:
                        new_grid[y][x] = INFECTED
        return new_grid

    def visualize_grid(grid, step):
        plt.figure(figsize=(5,4))
        plt.imshow(grid, cmap="viridis", interpolation="nearest")
        plt.colorbar(ticks=[HEALTHY, INFECTED, RECOVERED], 
                     format=plt.FuncFormatter(lambda val, loc: {HEALTHY: 'Healthy', INFECTED: 'Infected', RECOVERED: 'Recovered'}[val]))
        plt.title(f"COVID-19 Spread - Time Step {step}")
        plt.show()

    def count_infected_cells(grid):
        return np.sum(grid == INFECTED)

    def main():
        grid = initialize_grid()
        print("Covid spread with", i, "infection rate")

        for step in range(NUM_TIME_STEPS):
            print("Time Step:", step)

            # Count the infected cells
            count_infected_cells_var = count_infected_cells(grid)
            lst.append(count_infected_cells_var)

            # Print the count
            print("Infected Cells:", count_infected_cells_var)

            visualize_grid(grid, step)
            grid = update_grid(grid)

    if __name__ == "__main__":
        main()
```
